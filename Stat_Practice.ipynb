##CHI square Test

```python
import scipy.stats as stats
```

```python
expectd_data = [8,6,7,9,6,9,7]
observed_data = [7,8,6,9,9,6,7]
```

```python
sum(expectd_data),sum(observed_data)
```

```python
chisquare_test_statistic,p_value=stats.chisquare(observed_data,expectd_data)
```

```python
chisquare_test_statistic,p_value
```

#To find the critical value

```python
significance_value = 0.05
dof =len(expectd_data)-1
print(dof)
critical_value = stats.chi2.ppf(1-significance_value,dof)
```

```python
critical_value
```

```python
if chisquare_test_statistic > critical_value:
  print("Reject the null hypothesis")
else:
  print("Accept the null hypothesis")
```
