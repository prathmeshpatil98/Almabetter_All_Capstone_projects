## Normalization ( Min - Max Scaler )

```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit(df[['distance','fare','tip']])
```

```python
min_max.transform(df[['distance','fare','tip']])
```

```python
min_max.fit_transform(df[['distance','fare','tip']])
```

```python
min_max.transform([[1,3,4]])
```

## Unit Vector

```python
df1 = sns.load_dataset('iris')
df1
```

```python
from sklearn.preprocessing import normalize
```

```python
df1.columns
```

```python
normalize(df1[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
import pandas as pd
pd.DataFrame(normalize(df1[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]),
             columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```
